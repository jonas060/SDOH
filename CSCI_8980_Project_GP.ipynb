```python
hpsa_pc_full = pd.read_excel("Data/BCD_HPSA_FCT_DET_PC.xlsx") # Health Professional Shortage Area - Primary Care
# hpsa_dh_full = pd.read_csv("BCD_HPSA_FCT_DET_DH.csv") # Health Professional Shortage Area - Dental Health
# hpsa_dh_full.head()
hpsa_pc_full.head()
```

Keep if HPSA Designation Date is before 12/31/2017 and the Withdrawn Date is after 01/01/2016.

What is a geographic HPSA?
A shortage of providers for an entire group of people within a defined geographic area

What is a population HPSA?
A shortage of providers for a specific group of people within a defined geographic area. Examples include low-income populations, homeless populations, and migrant farmworker populations.

```python
hpsa_pc_condensed = hpsa_pc_full[(pd.to_datetime(hpsa_pc_full['HPSA Designation Date']) <= '2017-12-31')] # Need to pick one date to avoid duplicates.
hpsa_pc_condensed = hpsa_pc_condensed[(pd.to_datetime(hpsa_pc_condensed['Withdrawn Date']) > '2017-12-31') | hpsa_pc_condensed['Withdrawn Date'].isna()]
hpsa_pc_condensed = hpsa_pc_condensed[hpsa_pc_condensed['HPSA Geography Identification Number'] != 'POINT'] # Not specific locations (eg correctional facilities)
hpsa_pc_condensed = hpsa_pc_condensed[hpsa_pc_condensed['Designation Type'] == 'Geographic HPSA'] # Only when everybody in a specifc area has a shortage

## Remove columns - some might be useful but will want to get the demographic info from another source since we will have so much data manipulation here.
# columns_to_remove = ['Rural Status Code', 'HPSA Shortage', 'HPSA Provider Ratio Goal', 'HPSA Formal Ratio', 'HPSA Name', 'HPSA ID', 'Designation Type', 'HPSA FTE', 'HPSA Degree of Shortage', 'State and County Federal Information Processing Standard Code', 'State FIPS Code', 'HPSA Estimated Served Population', 'Common State County FIPS Code', 'Primary State FIPS Code', 'County or County Equivalent Federal Information Processing Standard Code', 'HPSA Estimated Underserved Population', 'Metropolitan Indicator', 'Common State FIPS Code', 'Rural Status', 'HPSA Component State Abbreviation', 'HPSA Designation Population', 'HPSA Designation Population Type Description', 'HPSA Component Source Identification Number', 'HPSA Postal Code', '% of Population Below 100% Poverty', 'BHCMIS Organization Identification Number', 'Common State Abbreviation', 'HPSA Population Type', 'Common Region Name', 'Common Postal Code', 'HPSA Resident Civilian Population', 'HPSA Type Code', 'HPSA Status', 'Primary State Abbreviation', 'State Abbreviation', 'State Name', 'HPSA Status Code', 'HPSA Withdrawn Date String', 'HPSA Discipline Class', 'HPSA Designation Last Update Date', 'Longitude', 'Latitude', 'Break in Designation', 'Common County Name', 'County Equivalent Name', 'HPSA Address', 'HPSA City', 'HPSA Component Name', 'HPSA Component Type Description', 'Common State Name', 'Discipline Class Number', 'Primary State Name', 'Provider Type', 'Data Warehouse Record Create Date', 'Data Warehouse Record Create Date Text', 'Unnamed: 65']
# hpsa_pc_condensed = hpsa_pc_condensed.drop(columns=columns_to_remove)
hpsa_pc_condensed = hpsa_pc_condensed.loc[:, ['HPSA Score', 'PC MCTA Score', 'HPSA Geography Identification Number', 'HPSA Component Type Code', 'HPSA Designation Date']]

# Sorting data by designation date. If there are duplicates, keeping the older record since that is what is more likely to affect the data?
hpsa_pc_condensed = hpsa_pc_condensed.sort_values(by='HPSA Designation Date')
hpsa_pc_condensed = hpsa_pc_condensed.drop_duplicates(subset='HPSA Geography Identification Number', keep='first')

hpsa_pc_condensed = hpsa_pc_condensed.drop(columns=['HPSA Designation Date'])
hpsa_pc_condensed
```

```python
hpsa_pc_cty = hpsa_pc_condensed[hpsa_pc_condensed['HPSA Component Type Code'] == 'SCTY'].drop(columns=['HPSA Component Type Code'])
hpsa_pc_csd = hpsa_pc_condensed[hpsa_pc_condensed['HPSA Component Type Code'] == 'CSD'].drop(columns=['HPSA Component Type Code']) #Tricky here because some areas are defined by CSD and others CT or CTY.
hpsa_pc_ct = hpsa_pc_condensed[hpsa_pc_condensed['HPSA Component Type Code'] == 'CT'].drop(columns=['HPSA Component Type Code'])

hpsa_pc_csd_with_ct = hpsa_pc_csd.merge(csd_to_ct_2010, left_on='HPSA Geography Identification Number', right_on='csd', how='inner')
hpsa_pc_ct_copy = hpsa_pc_ct[['HPSA Geography Identification Number']].rename(columns={'HPSA Geography Identification Number': 'HPSA Geography Identification Number_ct'})
# hpsa_pc_ct_copy
merge_csd_ct = hpsa_pc_csd_with_ct.merge(hpsa_pc_ct_copy, left_on='ct', right_on='HPSA Geography Identification Number_ct', how='left', indicator=True)

merge_csd_ct = merge_csd_ct[(merge_csd_ct['_merge'] != 'both')] # Drop records where the crosswalked census tract is already in the census tract dataframe
merge_csd_ct['HPSA Geography Identification Number'] = merge_csd_ct['ct'] # Update 'HPSA Geography Identification Number' to CT

merge_csd_ct.drop(columns=['county', 'csd', 'ct', 'HPSA Geography Identification Number_ct', '_merge'], inplace=True)
merge_csd_ct = merge_csd_ct.drop_duplicates()

merge_csd_ct_uniques = merge_csd_ct.drop_duplicates(subset='HPSA Geography Identification Number', keep=False, ignore_index=True)
merge_csd_ct_duplicates = merge_csd_ct[merge_csd_ct.duplicated(subset='HPSA Geography Identification Number', keep=False)].groupby(['HPSA Geography Identification Number']).agg({'HPSA Score': 'mean', 'PC MCTA Score': 'mean'}) # When the CT is duplicated, take the mean of the other scores.
merge_csd_ct_duplicates.reset_index(inplace=True)
merge_csd_ct = pd.concat([merge_csd_ct_uniques, merge_csd_ct_duplicates], ignore_index=True)

hpsa_pc_cty_with_ct = hpsa_pc_cty.merge(csd_to_ct_2010, left_on='HPSA Geography Identification Number', right_on='county', how='inner')

merge_csd_ct_copy = merge_csd_ct[['HPSA Geography Identification Number']].rename(columns={'HPSA Geography Identification Number': 'HPSA Geography Identification Number_ct'})
merge_cty_ct = hpsa_pc_cty_with_ct.merge(merge_csd_ct_copy, left_on='ct', right_on='HPSA Geography Identification Number_ct', how='left', indicator=True)

merge_cty_ct = merge_cty_ct[(merge_cty_ct['_merge'] != 'both')] # Drop records where the crosswalked census tract is already in the census tract dataframe
merge_cty_ct['HPSA Geography Identification Number'] = merge_cty_ct['ct'] # Update 'HPSA Geography Identification Number' to CT

merge_cty_ct.drop(columns=['county', 'csd', 'ct', 'HPSA Geography Identification Number_ct', '_merge'], inplace=True)
merge_cty_ct = merge_cty_ct.drop_duplicates()

merge_cty_ct_uniques = merge_cty_ct.drop_duplicates(subset='HPSA Geography Identification Number', keep=False, ignore_index=True)
merge_cty_ct_duplicates = merge_cty_ct[merge_cty_ct.duplicated(subset='HPSA Geography Identification Number', keep=False)].groupby(['HPSA Geography Identification Number']).agg({'HPSA Score': 'mean', 'PC MCTA Score': 'mean'}) # When the CT is duplicated, take the mean of the other scores.
merge_cty_ct_duplicates.reset_index(inplace=True)
merge_cty_ct = pd.concat([merge_cty_ct_uniques, merge_cty_ct_duplicates], ignore_index=True)
merge_all_hpsa = pd.concat([hpsa_pc_ct, merge_csd_ct, merge_cty_ct])
merge_all_hpsa['HPSA Geography Identification Number'] = merge_all_hpsa['HPSA Geography Identification Number'].astype('int64')
merge_all_hpsa.dtypes
# len(set(merge_all['HPSA Geography Identification Number']))
```

```python
merge_all_data = pd.merge(ct_500_cities.loc[:, ['MeasureId', 'TractFIPS', 'Data_Value']], cdc_svi_2016, left_on=['TractFIPS'], right_on=['FIPS'], how='inner')
merge_all_data = pd.merge(merge_all_data, merge_all_hpsa, left_on=['TractFIPS'], right_on=['HPSA Geography Identification Number'], how='left')
# merged_imputed = pd.merge(ct_500_cities.loc[:, ['MeasureId', 'TractFIPS', 'Data_Value']], cdc_svi_2016_imputed, left_on=['TractFIPS'], right_on=['FIPS'], how='inner')
merge_all_data[['HPSA Score', 'PC MCTA Score']] = merge_all_data[['HPSA Score', 'PC MCTA Score']].fillna(0)
# merge_data
merge_all_data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.model_selection import cross_validate
import matplotlib.pyplot as plt

measure_list = []
score_list = []
lin_reg_list = []
analysis_dict = {}

for measure in set(ct_500_cities['MeasureId']):
  merge_data_measure = merge_all_data[merge_all_data['MeasureId'] == measure].dropna(subset=['Data_Value'])

  scaler = StandardScaler()

  # X = diabetes.loc[:, diabetes.columns != 'Data_Value']
  y = merge_data_measure['Data_Value']
  # y_imputed = diabetes_imputed['Data_Value']

  # X_numeric = X.select_dtypes(include='number')
  columns_to_remove = ['Data_Value', 'TractFIPS', 'FIPS', 'F_TOTAL', 'HPSA Geography Identification Number']
  columns_to_remove = columns_to_remove + [col for col in merge_data_measure.columns if 'THEME' in col]
  columns_to_remove = columns_to_remove + [col for col in merge_data_measure.columns if col.startswith('E_')]
  columns_to_remove = columns_to_remove + [col for col in merge_data_measure.columns if col.startswith('M_')]
  columns_to_remove = columns_to_remove + [col for col in merge_data_measure.columns if col.startswith('EPL_')]
  columns_to_remove = columns_to_remove + [col for col in merge_data_measure.columns if col.startswith('MP_')]
  columns_to_remove = columns_to_remove + [col for col in merge_data_measure.columns if col.startswith('F_')]

  X_numeric = merge_data_measure.select_dtypes(include='number').drop(columns=columns_to_remove)
  X_numeric_scaled = scaler.fit_transform(X_numeric)

  linear_reg = LinearRegression().fit(X_numeric_scaled,y)

  measure_list.append(measure)
  score_list.append(linear_reg.score(X_numeric_scaled,y))
  lin_reg_list.append(linear_reg)

  # analysis_dict[measure] = (linear_reg.score(X_numeric_scaled,y), linear_reg.coef_)

  measure_desc = y.describe()
  bin_edges = [measure_desc[3], measure_desc[4], measure_desc[6], measure_desc[7]]
  bin_labels = ['low', 'medium', 'high']

  y_binned = pd.cut(y, bins=bin_edges, labels=bin_labels, include_lowest=True)

  X_rf = X_numeric_scaled
  num_trees = [50] # [1, 50, 100, 200, 300, 500, 1000]
  m =  X_rf.shape[1]
  max_features = ['sqrt'] #[ m, m//2, 'sqrt']
  test_accs = []
  feature_importances = []

  # Create StratifiedKFold validation
  n_splits = 10
  skf = StratifiedKFold(n_splits=n_splits, shuffle=True)

  for features in max_features:
      for tree in num_trees:
          # Instantiate a random forest object. Set max_depth=5.
          rf_clf = RandomForestClassifier(n_estimators=tree, max_features=features, max_depth=5)
          # rf_clf = RandomForestClassifier(n_estimators=tree, max_features=features, max_depth=10)

          # Perform cross validation
          rf_cv = cross_validate(rf_clf, X_rf, y_binned, scoring='accuracy', cv=skf, return_estimator=True)

          # Record the mean of CV test scores

          test_accs.append(np.mean(rf_cv['test_score']))

          importances = np.mean([estimator.feature_importances_ for estimator in rf_cv['estimator']], axis=0)
          feature_importances.append(importances)

  feature_names = X_numeric.columns

  # Display feature importances
  mean_feature_importances = np.mean(feature_importances, axis=0)
  std_feature_importances = np.std(feature_importances, axis=0)
  indices = np.argsort(mean_feature_importances)[::-1]
  indices

  feature_ranking = []

  # print("Feature ranking:")
  for f in range(len(feature_names)):
    feature_ranking.append("%d. feature %d (%f) - %s" % (f + 1, indices[f], mean_feature_importances[indices[f]], feature_names[indices[f]]))
      # print("%d. feature %d (%f) - %s" % (f + 1, indices[f], mean_feature_importances[indices[f]], feature_names[indices[f]]))

  analysis_dict[measure] = (linear_reg.score(X_numeric_scaled,y), linear_reg.coef_, test_accs, feature_ranking)

  # X_numeric_imputed = diabetes_imputed.select_dtypes(include='number').drop(columns=columns_to_remove)
  # X_numeric_scaled_imputed = scaler.fit_transform(X_numeric_imputed)

  # linear_reg_imputed = LinearRegression().fit(X_numeric_scaled_imputed,y)
  # X_numeric_imputed.head()
  # X_numeric_scaled
```
