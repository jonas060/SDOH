```python
diabetes = merged[merged['MeasureId'] == 'DIABETES'].dropna(subset=['Data_Value'])
# diabetes_imputed = merged_imputed[merged_imputed['MeasureId'] == 'DIABETES'].dropna(subset=['Data_Value'])
# diabetes.head()

print(diabetes['Data_Value'].describe())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# X = diabetes.loc[:, diabetes.columns != 'Data_Value']
y = diabetes['Data_Value']
# y_imputed = diabetes_imputed['Data_Value']

# X_numeric = X.select_dtypes(include='number')
columns_to_remove = ['Data_Value', 'TractFIPS', 'FIPS', 'F_TOTAL'] #, 'RPL_THEMES', 'SPL_THEMES', 'F_THEME1', 'RPL_THEME1', 'SPL_THEME1', 'F_THEME2', 'RPL_THEME2', 'SPL_THEME2', 'F_THEME3', 'RPL_THEME3', 'SPL_THEME3', 'F_THEME4', 'RPL_THEME4', 'SPL_THEME4']
columns_to_remove = columns_to_remove + [col for col in diabetes.columns if 'THEME' in col]
columns_to_remove = columns_to_remove + [col for col in diabetes.columns if col.startswith('E_')]
columns_to_remove = columns_to_remove + [col for col in diabetes.columns if col.startswith('M_')]
columns_to_remove = columns_to_remove + [col for col in diabetes.columns if col.startswith('EPL_')]
columns_to_remove = columns_to_remove + [col for col in diabetes.columns if col.startswith('MP_')]
columns_to_remove = columns_to_remove + [col for col in diabetes.columns if col.startswith('F_')]

X_numeric = diabetes.select_dtypes(include='number').drop(columns=columns_to_remove)
X_numeric_scaled = scaler.fit_transform(X_numeric)

linear_reg = LinearRegression().fit(X_numeric_scaled,y)
# X_numeric.head()
# columns_to_remove

# X_numeric_imputed = diabetes_imputed.select_dtypes(include='number').drop(columns=columns_to_remove)
# X_numeric_scaled_imputed = scaler.fit_transform(X_numeric_imputed)

# linear_reg_imputed = LinearRegression().fit(X_numeric_scaled_imputed,y)
# X_numeric_imputed.head()
```

```python
diabetes_desc = diabetes['Data_Value'].describe()
bin_edges = [diabetes_desc[3], diabetes_desc[4], diabetes_desc[6], diabetes_desc[7]]
bin_labels = ['low', 'medium', 'high']

y_binned = pd.cut(y, bins=bin_edges, labels=bin_labels, include_lowest=True)

print(y_binned)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.model_selection import cross_validate
import matplotlib.pyplot as plt

X_rf = X_numeric_scaled
# X_rf_2 = X_projected[:, 0:25]
num_trees = [1, 50, 100, 200, 300, 500] # [1, 50, 100, 200, 300, 500, 1000]
m =  X_rf.shape[1]
# m_2 = X_rf_2.shape[1]
max_features = [ m, m//2, 'sqrt']
test_accs = []

# Create StratifiedKFold validation
n_splits = 10
skf = StratifiedKFold(n_splits=n_splits, shuffle=True)

for features in max_features:
    for tree in num_trees:
        # Instantiate a random forest object. Set max_depth=5.
        # rf_clf = RandomForestClassifier(n_estimators=tree, max_features=features, max_depth=5)
        rf_clf = RandomForestClassifier(n_estimators=tree, max_features=features, max_depth=10)

        # Perform cross validation
        rf_cv = cross_validate(rf_clf, X_rf, y_binned, scoring='accuracy', cv=skf, return_estimator=True)

        # Record the mean of CV test scores

        test_accs.append(np.mean(rf_cv['test_score']))


# print(test_accs)
# print(test_accs[1])

plt.xlabel('Number of Trees')
plt.ylabel('CV test accuracy')
for i, max_feat in enumerate(max_features):
    plt.plot(num_trees, test_accs[i*len(num_trees):(i*len(num_trees))+len(num_trees)], label="max_features = "+str(max_feat))
plt.legend(loc="lower right")

plt.show()
```
