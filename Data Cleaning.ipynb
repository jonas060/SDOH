```python
import pandas as pd
import numpy as np
```

# Import and Clean Data

## 500 Cities Data

```python
places_500_cities = pd.read_csv("Data/500_Cities__Local_Data_for_Better_Health__2019_release_20240314.zip", dtype={'TractFIPS': str})
places_500_cities.loc[:, ['Measure', 'MeasureId']].drop_duplicates()
ct_500_cities = places_500_cities[places_500_cities['GeographicLevel'] == 'Census Tract']
columns_to_remove = ['StateDesc', 'GeographicLevel', 'DataSource', 'UniqueID', 'Measure', 'Low_Confidence_Limit', 'High_Confidence_Limit', 'GeoLocation', 'CategoryID', 'Short_Question_Text']
ct_500_cities = ct_500_cities.drop(columns=columns_to_remove)
ct_500_cities
```

## SVI 2106 Census Tracts

```python
cdc_svi_2016_full = pd.read_csv("Data/SVI_2016_Census_Tracts.csv", dtype={'FIPS': str})

cdc_svi_2016_full = cdc_svi_2016_full[cdc_svi_2016_full['E_TOTPOP'] != 0] ## Removing rows where the E_TOTPOP = 0
cdc_svi_2016_full['POP_DENSITY'] = cdc_svi_2016_full['E_TOTPOP']/cdc_svi_2016_full['AREA_SQMI']

columns_to_remove = ['ST', 'STATE', 'ST_ABBR', 'STCNTY', 'COUNTY', 'LOCATION']
columns_to_remove = columns_to_remove + [col for col in cdc_svi_2016_full.columns if 'THEME' in col]
columns_to_remove = columns_to_remove + [col for col in cdc_svi_2016_full.columns if col.startswith('M_')]
columns_to_remove = columns_to_remove + [col for col in cdc_svi_2016_full.columns if col.startswith('MP_')]
columns_to_remove = columns_to_remove + [col for col in cdc_svi_2016_full.columns if col.startswith('E_')]
columns_to_remove = columns_to_remove + [col for col in cdc_svi_2016_full.columns if col.startswith('EPL_')]
columns_to_remove = columns_to_remove + [col for col in cdc_svi_2016_full.columns if col.startswith('F_')]
cdc_svi_2016 = cdc_svi_2016_full.drop(columns=columns_to_remove)

cdc_svi_2016
```

## HSPA (Health Professional Shortage Area) Data

Keep if HPSA Designation Date is on or before 12/31/2017 and the Withdrawn Date is after 12/31/2017.

What is a geographic HPSA?
A shortage of providers for an entire group of people within a defined geographic area

What is a population HPSA?
A shortage of providers for a specific group of people within a defined geographic area. Examples include low-income populations, homeless populations, and migrant farmworker populations.

```python
hpsa_pc_full = pd.read_csv("Data/BCD_HPSA_FCT_DET_PC.csv") # Health Professional Shortage Area - Primary Care
hpsa_dh_full = pd.read_csv("Data/BCD_HPSA_FCT_DET_DH.csv") # Health Professional Shortage Area - Dental Health
hpsa_pc_dh_full = pd.merge(hpsa_pc_full, hpsa_dh_full, how='outer')

hpsa_pc_dh = hpsa_pc_dh_full[(pd.to_datetime(hpsa_pc_dh_full['HPSA Designation Date']) <= '2017-12-31')] # Need to pick one date to avoid duplicates.
hpsa_pc_dh = hpsa_pc_dh[(pd.to_datetime(hpsa_pc_dh['Withdrawn Date']) > '2017-12-31') | hpsa_pc_dh['Withdrawn Date'].isna()]
hpsa_pc_dh = hpsa_pc_dh[hpsa_pc_dh['HPSA Geography Identification Number'] != 'POINT'] # Not specific locations (eg correctional facilities)
hpsa_pc_dh = hpsa_pc_dh[hpsa_pc_dh['Designation Type'] == 'Geographic HPSA'] # Only when everybody in a specifc area has a shortage

## Remove columns - some might be useful but will want to get the demographic info from another source since we will have so much data manipulation here.
hpsa_pc_dh = hpsa_pc_dh.loc[:, ['HPSA Discipline Class', 'HPSA Score', 'PC MCTA Score', 'HPSA Geography Identification Number', 'HPSA Component Type Code', 'HPSA Designation Date']]

# Sorting data by designation date. If there are duplicates, keeping the older record since that is what is more likely to affect the data?
hpsa_pc_dh = hpsa_pc_dh.sort_values(by='HPSA Designation Date')
hpsa_pc_dh = hpsa_pc_dh.drop_duplicates(subset='HPSA Geography Identification Number', keep='first')


hpsa_pc_dh['PC_HPSA_Score'] = np.where(hpsa_pc_dh['HPSA Discipline Class'] == 'Primary Care', hpsa_pc_dh['HPSA Score'], np.nan)
hpsa_pc_dh['DH_HPSA_Score'] = np.where(hpsa_pc_dh['HPSA Discipline Class'] == 'Dental Health', hpsa_pc_dh['HPSA Score'], np.nan)

hpsa_pc_dh = hpsa_pc_dh.drop(columns=['HPSA Designation Date', 'HPSA Score', 'HPSA Discipline Class'])
hpsa_pc_dh
```

## Demographic Housing Data

```python
acs_demog_house_est = pd.read_csv("Data/Demographic_Housing_DP05-Data_CLEAN.csv", dtype={"Geography": str})

acs_demog_house_est = acs_demog_house_est.drop(columns=['Unnamed: 0'])
acs_demog_house_est = acs_demog_house_est.drop(columns=['Percent!!CITIZEN, VOTING AGE POPULATION!!Citizen, 18 and over population', 'Percent!!CITIZEN, VOTING AGE POPULATION!!Citizen, 18 and over population!!Male', 'Percent!!CITIZEN, VOTING AGE POPULATION!!Citizen, 18 and over population!!Female', 'Percent!!SEX AND AGE!!Median age (years)']) # Removing the fields with totals and not percents.

features = acs_demog_house_est.drop(columns='Geography')
bad_rows = features[((features == '-') | (features == 0)).all(axis=1)].index # Removing census tracts where all values are blank or 0
acs_demog_house_est =acs_demog_house_est.drop(bad_rows)

acs_demog_house_est['Percent!!SEX AND AGE!!65 years and over!!Male'] = acs_demog_house_est['Percent!!SEX AND AGE!!65 years and over!!Male'].replace('-', 50) # If these columns are blank (there was nobody age 65+ in this census tract), setting at 50%. Maybe it would be better to set as the overall male/female percentages.
acs_demog_house_est['Percent!!SEX AND AGE!!65 years and over!!Female'] = acs_demog_house_est['Percent!!SEX AND AGE!!65 years and over!!Female'].replace('-', 50)


for column in acs_demog_house_est.columns:
  if column.startswith('Percent!!'):
    acs_demog_house_est[column] = pd.to_numeric(acs_demog_house_est[column], errors='coerce')


acs_demog_house_est
```

## Disability Data

```python
disab_chars = pd.read_csv("Data/Disability_Characteristics_S1810-Data_CLEAN.csv", dtype={"Geography": str})

for column in disab_chars.columns:

  if column.startswith('Estimate!!Percent'):
      disab_chars[column] = pd.to_numeric(disab_chars[column], errors='coerce')
      disab_chars[column] = disab_chars[column].fillna(0) # I might be making too many assumptions here. Look into this more. Impute?

disab_chars = disab_chars.drop(columns='Unnamed: 0')
disab_chars
```

## Match Tract FIPS With Subcounty FIPS

```python
csd_to_ct_2010 = pd.read_csv("Data/CSD_to_CT_2010.csv", dtype=str)
csd_to_ct_2010['csd'] = csd_to_ct_2010['county'] + csd_to_ct_2010['cousubfp']
csd_to_ct_2010['tract'] = csd_to_ct_2010['tract'].apply(lambda x: x + '00' if '.' not in x else x)
csd_to_ct_2010['ct'] = csd_to_ct_2010['county'] + csd_to_ct_2010['tract'].str.replace(".", "")

csd_to_ct_2010
```

# Merge Data Based On FIPS

```python
hpsa_cty = hpsa_pc_dh[hpsa_pc_dh['HPSA Component Type Code'] == 'SCTY'].drop(columns=['HPSA Component Type Code'])
hpsa_csd = hpsa_pc_dh[hpsa_pc_dh['HPSA Component Type Code'] == 'CSD'].drop(columns=['HPSA Component Type Code']) #Tricky here because some areas are defined by CSD and others CT or CTY.
hpsa_ct = hpsa_pc_dh[hpsa_pc_dh['HPSA Component Type Code'] == 'CT'].drop(columns=['HPSA Component Type Code'])

hpsa_csd_with_ct = hpsa_csd.merge(csd_to_ct_2010, left_on='HPSA Geography Identification Number', right_on='csd', how='inner')
hpsa_ct_copy = hpsa_ct[['HPSA Geography Identification Number']].rename(columns={'HPSA Geography Identification Number': 'HPSA Geography Identification Number_ct'})
# hpsa_pc_ct_copy
merge_csd_ct = hpsa_csd_with_ct.merge(hpsa_ct_copy, left_on='ct', right_on='HPSA Geography Identification Number_ct', how='left', indicator=True)

merge_csd_ct = merge_csd_ct[(merge_csd_ct['_merge'] != 'both')] # Drop records where the crosswalked census tract is already in the census tract dataframe
merge_csd_ct['HPSA Geography Identification Number'] = merge_csd_ct['ct'] # Update 'HPSA Geography Identification Number' to CT

merge_csd_ct.drop(columns=['county', 'csd', 'ct', 'HPSA Geography Identification Number_ct', '_merge'], inplace=True)
merge_csd_ct = merge_csd_ct.drop_duplicates()

merge_csd_ct_uniques = merge_csd_ct.drop_duplicates(subset='HPSA Geography Identification Number', keep=False, ignore_index=True)
merge_csd_ct_duplicates = merge_csd_ct[merge_csd_ct.duplicated(subset='HPSA Geography Identification Number', keep=False)].groupby(['HPSA Geography Identification Number']).agg({'PC_HPSA_Score': 'mean', 'DH_HPSA_Score': 'mean', 'PC MCTA Score': 'mean'}) # When the CT is duplicated, take the mean of the other scores.
merge_csd_ct_duplicates.reset_index(inplace=True)
merge_csd_ct = pd.concat([merge_csd_ct_uniques, merge_csd_ct_duplicates], ignore_index=True)

hpsa_cty_with_ct = hpsa_cty.merge(csd_to_ct_2010, left_on='HPSA Geography Identification Number', right_on='county', how='inner')

merge_csd_ct_copy = merge_csd_ct[['HPSA Geography Identification Number']].rename(columns={'HPSA Geography Identification Number': 'HPSA Geography Identification Number_ct'})
merge_cty_ct = hpsa_cty_with_ct.merge(merge_csd_ct_copy, left_on='ct', right_on='HPSA Geography Identification Number_ct', how='left', indicator=True)

merge_cty_ct = merge_cty_ct[(merge_cty_ct['_merge'] != 'both')] # Drop records where the crosswalked census tract is already in the census tract dataframe
merge_cty_ct['HPSA Geography Identification Number'] = merge_cty_ct['ct'] # Update 'HPSA Geography Identification Number' to CT

merge_cty_ct.drop(columns=['county', 'csd', 'ct', 'HPSA Geography Identification Number_ct', '_merge'], inplace=True)
merge_cty_ct = merge_cty_ct.drop_duplicates()

merge_cty_ct_uniques = merge_cty_ct.drop_duplicates(subset='HPSA Geography Identification Number', keep=False, ignore_index=True)
merge_cty_ct_duplicates = merge_cty_ct[merge_cty_ct.duplicated(subset='HPSA Geography Identification Number', keep=False)].groupby(['HPSA Geography Identification Number']).agg({'PC_HPSA_Score': 'mean', 'DH_HPSA_Score': 'mean', 'PC MCTA Score': 'mean'}) # When the CT is duplicated, take the mean of the other scores.
merge_cty_ct_duplicates.reset_index(inplace=True)
merge_cty_ct = pd.concat([merge_cty_ct_uniques, merge_cty_ct_duplicates], ignore_index=True)
merge_all_hpsa = pd.concat([hpsa_ct, merge_csd_ct, merge_cty_ct])
merge_all_hpsa['HPSA Geography Identification Number'] = merge_all_hpsa['HPSA Geography Identification Number']

merge_all_hpsa.head()
```

## Training Data (include 500 cities)

```python
merge_all_data_training = pd.merge(ct_500_cities.loc[:, ['MeasureId', 'TractFIPS', 'Data_Value']], cdc_svi_2016, left_on=['TractFIPS'], right_on=['FIPS'], how='inner')
merge_all_data_training = pd.merge(merge_all_data_training, merge_all_hpsa, left_on=['TractFIPS'], right_on=['HPSA Geography Identification Number'], how='left')
merge_all_data_training = pd.merge(merge_all_data_training, acs_demog_house_est, left_on=['TractFIPS'], right_on=['Geography'], how='left')
merge_all_data_training = pd.merge(merge_all_data_training, disab_chars, left_on=['TractFIPS'], right_on=['Geography'], how='left')
merge_all_data_training[['PC_HPSA_Score', 'DH_HPSA_Score', 'PC MCTA Score']] = merge_all_data_training[['PC_HPSA_Score', 'DH_HPSA_Score', 'PC MCTA Score']].fillna(0) # If NaN, these scores weren't on files. This means they aren't considered an HPSA. So a score of 0.


merge_all_data_training = merge_all_data_training.drop_duplicates(subset=['FIPS'])
merge_all_data_training = merge_all_data_training.drop(columns=['Geography_x', 'Geography_y', 'TractFIPS', 'HPSA Geography Identification Number', 'cousubfp', 'tract'])
merge_all_data_training
```

## Testing Data (exclude 500 cities)

```python
merge_all_data_testing = pd.merge(cdc_svi_2016, merge_all_hpsa, left_on=['FIPS'], right_on=['HPSA Geography Identification Number'], how='left')
merge_all_data_testing = pd.merge(merge_all_data_testing, acs_demog_house_est, left_on=['FIPS'], right_on=['Geography'], how='left')
merge_all_data_testing = pd.merge(merge_all_data_testing, disab_chars, left_on=['FIPS'], right_on=['Geography'], how='left')
merge_all_data_testing[['PC_HPSA_Score', 'DH_HPSA_Score', 'PC MCTA Score']] = merge_all_data_testing[['PC_HPSA_Score', 'DH_HPSA_Score', 'PC MCTA Score']].fillna(0) # If NaN, these scores weren't on files. This means they aren't considered an HPSA. So a score of 0.


merge_all_data_testing = merge_all_data_testing.drop_duplicates(subset=['FIPS'])
merge_all_data_testing = merge_all_data_testing.drop(columns=['Geography_x', 'Geography_y', 'HPSA Geography Identification Number', 'cousubfp', 'tract'])
merge_all_data_testing
```
